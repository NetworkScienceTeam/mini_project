```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.karate_club_graph()

# Calcular la centralidad
centrality = nx.betweenness_centrality(G)
print("Betweenness Centrality")
for v in centrality:
    print(f" {v:4} {centrality[v]:0.3f}")

# Calcular la modularidad
communities = nx.algorithms.community.greedy_modularity_communities(G)
modularity = nx.algorithms.community.modularity(G, communities)
print(f"Modularity: {modularity:0.3f}")

# Calcular las comunidades
print("Communities")
for i, community in enumerate(communities):
    print(f" Community {i+1}: {community}")
```

```python
sorted(centrality.items(), key=lambda x:x[1], reverse = True)
```

```python
nx.draw_circular(G, with_labels=True)
plt.show()
```

```python
# Detectar las comunidades
partition = nx.algorithms.community.greedy_modularity_communities(G)

# Dibujar la red separada por comunidades
pos = nx.circular_layout(G)
for i, com in enumerate(communities):
    nx.draw_networkx_nodes(G, pos, com, node_size=200, node_color=f"C{i}")
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos)
plt.show()
```

```python
pos = nx.spring_layout(G)
for i, com in enumerate(communities):
    nx.draw_networkx_nodes(G, pos, com, node_size=200, node_color=f"C{i}")
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos)
plt.show()
```

```python
# Calcular el grado de cada nodo
degree = dict(G.degree())

# Calcular la centralidad de grado de cada nodo
degree_centrality = nx.degree_centrality(G)

# Detectar las comunidades
partition = nx.algorithms.community.greedy_modularity_communities(G)

# Dibujar la red separada por comunidades con una distribucin diferente
pos = nx.spring_layout(G)
for i, com in enumerate(communities):
    nx.draw_networkx_nodes(G, pos, com, node_size=200, node_color=f"C{i}")
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos)

# Mostrar los valores de grado, centralidad de grado y comunidad de cada nodo
for node in G.nodes():
  print(f"Nodo {node}: grado = {degree[node]}, centralidad de grado = {degree_centrality[node]:.2f}")
  for comunidad in partition:
    if node in comunidad:
        print(f"El nodo {node} pertenece a la comunidad {comunidad}.")

plt.show()
```
